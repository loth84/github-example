```python
# Import different library
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate
from IPython.display import HTML, display
```

```python
# Import librairy
from IPython.display import HTML, display
import tabulate
```

```python
# Code for import the data frame and the table contant in the page
res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))[0]
#print( tabulate(df[0], headers='keys', tablefmt='psql') )
df[0:3]
```

### Drop row with not assigmed 

```python
#Drop row with not assigmed
df = df[df.Borough != 'Not assigned']
df[0:5]
```

### Aggregate values [ https://kanoki.org/2019/09/04/pandas-groupby-tutorial/]  [ https://stackoverflow.com/questions/53781634/aggregation-in-pandas ]

```python
# marge only the Neighbourhod coll in dependancy of Postcode ans Borough
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].agg(','.join).reset_index()
df[0:3]
```

### Select row with value Not assigmed and change the cell

```python
# Select the row where the in Neighbourhood there is the value Not assigned and declare a new dataframe to show how the change is done 
df.loc[df['Neighbourhood'] == 'Not assigned']
df[0:86]      
```

```python
# eplacement of pandas cell values with cell values of another row and print the result
df.loc[df.Neighbourhood =='Not assigned', 'Neighbourhood']=df.Borough
df[0:86]
```
